```python
import numpy as np
import pandas as pd

# Menampilkan dataset / DataFrame
df = pd.read_excel('Sample_Superstore.xlsx')
df
```

```python
print("Menemukan Top 10 produk berdasarkan total Sales\n")
print("===============================================\n")

top_10 = df.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head(10)
top_10

```

```python
print("Kota yang memiliki profit tertinggi\n")
print("===============================================\n")

mean = df.groupby('City')['Profit'].mean().sort_values(ascending=False)

kota_tertinggi = mean.index[0]
kota_tertinggi
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("Bar Plot\n")
print("===============================================\n")

# Menghitung rata-rata profit per kota dan mengurutkan
rata_rata_profit_per_kota = df.groupby('City')['Profit'].mean().sort_values(ascending=False)

# Membuat bar plot
plt.bar(rata_rata_profit_per_kota.index, rata_rata_profit_per_kota.values)
plt.xlabel('Kota')
plt.ylabel('Rata-rata Profit')
plt.title('Rata-rata Profit per Kota')
plt.xticks(rotation=45)  # Memutar label sumbu x agar lebih mudah dibaca
plt.show()
```
