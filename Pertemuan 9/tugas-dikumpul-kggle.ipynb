```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score

#Import data
df = pd.read_csv('Mall_Customers2.csv')
df

# Memilih kolom untuk dilakukan clustering
data = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Clustering menggunakan KMeans
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(data)
df['KMeans_Cluster'] = kmeans.labels_

#Clustering mengunakan DBSCAN
dbscan = DBSCAN(eps=3, min_samples=5)
dbscan.fit(data)
df['DBSCAN_Cluster'] = dbscan.labels_

#Visualisasi hasil clustering
plt.figure(figsize=(12, 6))

#KMeans
plt.subplot(1, 2, 1)
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['KMeans_Cluster'], cmap='rainbow')
plt.title('K-Means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spanding Score (1-100)')

#DBSCAN
plt.subplot(1, 2, 2)
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['KMeans_Cluster'], cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spanding Score (1-100)')

plt.tight_layout()
plt.show
```
