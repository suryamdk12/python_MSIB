```python
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker  #Library untuk membuat grfik
import matplotlib.pyplot as plt
import seaborn as sns    


sns.set()
import warnings
warnings.filterwarnings('ignore')


print("\n=================================================")
print("INI ADALAH DATA DARI SAMPLE SUPERSTORE DATA CSV")
print("=================================================\n")
#Read SampelSuperstore data
df = pd.read_csv('SampleSuperstore.csv') #Read the dataset
df
```

```python
print("\n=================================================")
print("INI ADALAH DATA SAMPLE SUPERSTORE DATA CSV YNAG SUDAH DI HITUNG DESKRIPSI STATISTIK DARI DATA NUMAERIK")
print("=================================================\n")
#  Menghitung statistik deskriptif dari data numerik dalam DataFrame
data = df.describe(include="all")
data
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MEMBUAT DAN MENAMPILKAN HISTOGRAM DARI SEMUA KOLOM NUMERIK DALAM DATAFRAME")
print("=================================================\n")

# Fungsi ini digunakan untuk membuat dan menampilkan histogram dari semua kolom numerik dalam dataframe
print(df.hist(bins=10, figsize=(18, 10)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

print("\n=================================================")
print("FUNGSI INI DIPERUNTUKKAN UNTUK MENGUBAH DATA DARI DATAFRAME MENJADI DATA GRAFIK")
print("=================================================\n")
# Fungsi untuk mengubah data menjadi data grafik warna


plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.title('Matriks Korelasi')
plt.show()
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MEMBUAT MATRIKS PLOT YANG MENUNJUKKAN HUBUNGAN ANTARA SEMUA PASANGAN VARIABEL NUMERIK DALAM DATAFRAME")
print("=================================================\n")

# Fungsi ini digunaakan untuk membuat matriks plot
# Menunjukkan hubungan antara semua pasangan variabel numerik dalam dataframe
sns.pairplot(df, hue='Ship Mode')
```

```python
print("\n=================================================")
print("FUNGSI INI DIPERUNTUKKAN UNTUK MENGUBAH DATA DARI DATAFRAME MENJADI SHIP MODE BENTUK DIAGRAM BAR")
print("=================================================\n")

# fungsi ini digunakan untuk membuat diagram batang(bar plot)
# Menunjukkan frekuensi atau jumlah kemunculan setiap kategori dalam kolom 'Ship Mode'
# Ship mode dagram batang
sns.countplot(x='Ship Mode', data=df, palette='tab10') #sns.countplot untuk membuat diagram batang dari seaborn
```

```python
print("\n=================================================")
print("FUNGSI INI DIPERUNTUKKAN UNTUK MENGUBAH DATA DARI DATAFRAME MENJADI SHIP MODE BENTUK DIAGRAM BAR")
print("=================================================\n")

# Fungsi ini digunaan untuk mengelompokkan data berdasarkan nilai unik dalam kolom 'Ship Mode'ArithmeticError
# Data akan dibagi menjadi beberapa kelompok yang masing-masing mewakili satu jenis 'Ship Mode'

# Fungsi keseluruhan adalah digunakan untuk menghitung rata-rata penjualan, keuntungan dan diskon untuk setiap jenis mode pengiriman (ship mode) dalam dataset
df_shipmode = df.groupby(['Ship Mode']) [['Sales', 'Profit', 'Discount']].mean() #fungsi mean disini sebagai menghitung rata-rata dari kolom yang telah ditentukan untuk setiap kelompok
df_shipmode
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MEMBUAT DIAGRAM PAI (PIE CHART) YANG MENAMPILKAN FISTRIBUSI DATA DALAM DATAFRAME")
print("=================================================\n")

# 
colors_4 = ['magenta', 'yellow', 'orange', 'red']
colors_3 = ['green', 'blue', 'cyan']

df_shipmode.plot.pie(subplots=True, figsize=(20, 10), labels = df_shipmode.index, autopct = '%1.1f%%', explode = (0.01, 0.01, 0.01, 0.01), colors=colors_4)
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MEMBUAT PERBANDINGAN NILAI DALAM DIAGRAM BATANG")
print("=================================================\n")

# Fungsi ini digunakan untuk membuat visualisasi data dalam bentuk diagram batang 
# Secara spesifik : untuk membandingkan nilai 'Profit' dan 'Sales' untuk setiap kategori dalam dataFra

df_shipmode.sort_values('Profit')[['Profit', 'Sales']].plot(kind='bar', figsize=(10, 7))
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MENGHITUNG JUMLAH TOTAL DARI SETIAP KOLOM DALAM DATAFRAME")
print("=================================================\n")

# Fungsi ini digunakan untuk melakukan analisis
# Menghitung jumlah total dari setiap kolom
df_shipmode.sum(axis=0)
df_shipmode
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MENGHITUNG PRESENTASE 'SALES' DAN 'PROFIT' BERDASARKAN TOTAL TERTENTU")
print("=================================================\n")

# fungsi df_shipmode : menghitung presentase 'sales' dan 'profit'

# Membagi setiap nilai dalam kolom 'Sales' dengan 9.285. Hasilnya disimpan dalam kolom baru bernama 'Sales%'.
df_shipmode['Sales%'] = df_shipmode['Sales']/9.285

#Membagi setiap nilai dalam kolom 'Profit' dengan 1.181. Hasilnya disimpan dalam kolom baru bernama 'Profit%'.
df_shipmode['Profit%'] = df_shipmode['Profit']/1.181

# Mengurutkan DataFrame df_shipmode berdasarkan nilai kolom 'Profit%' secara menaik.
df_shipmode.sort_values('Profit%')
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MEMBUAT PAIR PLOT ATAU PLOT PASANGAN")
print("=================================================\n")

# igunakan untuk membuat pair plot atau plot pasangan
sns.pairplot(df, hue='Segment')
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MEMBUAT VISUALISASI DISTRIBUSI FREKUENSI KATEGORIKAL")
print("=================================================\n")

# untuk membuat count plot atau plot hitungan.
# memvisualisasikan distribusi frekuensi kategorikal dari suatu variabel dalam dataset
sns.countplot(x='Segment', data = df, palette='tab10')
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MEMBUAT DIAGRAM PIE DARI NILAI RATA-RATA KOLOM SALES, PROFIT DAN DISCOUNT")
print("=================================================\n")


df_segment = df.groupby(['Segment'])[['Sales', 'Profit', 'Discount']].mean() #Mengelompokkan DataFrame df berdasarkan kolom 'Segment'.
#[['Sales', 'Profit', 'Discount']]: Memilih hanya kolom 'Sales', 'Profit', dan 'Discount' dari setiap kelompok.
df_segment.plot.pie(subplots=True, figsize=(20,10), labels=df_segment.index, autopct='%1.1f%%', colors=colors_3)
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MEMBUAT VISUALISASI DATA DALAM BENTUK DAGRAM BATANG 9BAR CHART)")
print("=================================================\n")

df_segment.sort_values('Profit')[['Profit', 'Sales']].plot(kind='bar')
```

```python
print("\n=================================================")
print("FUNGSI INI DIGUNAKAN UNTUK MENJUMLAHKAN TOTAL PENJUALAN, PROFIT DAN SALES DARI KESELURUHAN")
print("=================================================\n")

# digunakan untuk mendapatkan total dari setiap kolom dalam DataFrame
df_segment.sum(axis=0)
df_segment
```
