```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
#Data penjualan minggu 1 sampai 5
x=[[1],[2],[3],[4],[5]] 
#x adalah minggu, sifat : independent
y=[[150], [157], [162], [166], [177]] 
#y adalah data penjualan di minggu tersebut. sifat : dependent
x_train=x
y_train=y
lm = LinearRegression()
lm.fit(x_train,y_train) #fase training
import pandas as pd
x_test = [[6]] #data yang akan diprediksi, penjualan ke 6
p = lm.predict(x_test) 
#fase prediksi
print('Angka ramalan untuk penjualan minggu ke 6 : ',p) 
#hasil prediksi
#prepare plot
pb = lm.predict(x_train)
plt.scatter(x_train,y_train)
plt.plot(x_train,pb,color='red',linewidth=2)
plt.xlabel('Minggu')
plt.ylabel('Penjualan')
plt.show()
```
